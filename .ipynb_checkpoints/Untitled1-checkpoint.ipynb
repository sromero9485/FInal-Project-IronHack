```python
data = pd.read_csv('data_weed.csv')
data.head()
```

```python
data.EFA.unique()
```

```python
import spacy
from spacy.tokenizer import Tokenizer
from spacy.lang.en import English

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.neighbors import NearestNeighbors

import pickle
```

```python
nlp=English()
tokenizer = Tokenizer(nlp.vocab)
```

```python
tf = TfidfVectorizer(stop_words='english')
```

```python
matrix1 = tf.fit_transform(data['EFA'].values.astype('U'))
matrix1 = pd.DataFrame(matrix1.todense(), columns=tf.get_feature_names())
matrix1.head()
```

```python
matrix1.shape
```

```python
from umap import UMAP
import matplotlib.pyplot as plt
```

```python
umap=UMAP(n_neighbors=10, metric='cosine')
emb=umap.fit_transform(matrix1)
```

```python
emb[:5]
```

```python
plt.scatter(emb[:,0], emb[:,1])
```

```python
from sklearn.metrics import silhouette_score


from IPython.display import set_matplotlib_formats
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
set_matplotlib_formats('svg')

from yellowbrick.cluster import KElbowVisualizer
from sklearn.cluster import KMeans
```

```python
modelo=KMeans()

visual=KElbowVisualizer(modelo, k=(2, 8))
visual.fit(emb)

visual.poof();
```

```python
modelo=KMeans()

visual=KElbowVisualizer(modelo, metric='silhouette', k=(2, 8))
visual.fit(emb)

visual.poof();
```

```python
modelo=KMeans()

visual=KElbowVisualizer(modelo, metric='calinski_harabasz', k=(2,8))
visual.fit(emb)

visual.poof();
```

```python

from sklearn.cluster import AgglomerativeClustering
model2 = AgglomerativeClustering()
visualizer = KElbowVisualizer(model2, k=(2,8))
visualizer.fit(emb)
visualizer.poof()
```

```python
kmeans = KMeans(n_clusters=3)
clusters = kmeans.fit(emb)
score = silhouette_score(emb, clusters.labels_, metric='cosine')
score
```

```python
len(clusters.cluster_centers_)
```

```python
clusters.cluster_centers_
```

```python
preds=clusters.fit_predict(emb)
```

```python
preds
```

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA().fit(matrix1)
pca_x=pca.transform(matrix1)
```

```python
fig=plt.figure(figsize=(12, 8))

ax=[plt.subplot2grid((6, 1), (0, 0), rowspan=4),
    plt.subplot2grid((6, 1), (5, 0))]

ax[0].scatter(pca_x[:,0], pca_x[:,1])
ax[0].set_xlabel('PCA 1')
ax[0].set_ylabel('PCA 2')

ax[1].set_title('PCA 3', pad=-5)
ax[1].hlines(1, -7, 10)

y=np.ones(matrix1.shape[0])

ax[1].plot(pca_x[:,2], y, '|', ms='20')
ax[1].axis('off')

plt.show()
```

```python
ig = plt.figure(figsize=(8, 8))
ax = [plt.subplot2grid((6, 1), (0, 0), rowspan=4)]

ax[0].scatter(pca_x[:, 0], pca_x[:, 1], cmap='Set1')
ax[0].set_xlabel("PCA 1")
ax[0].set_ylabel("PCA 2")

plt.show();
```
